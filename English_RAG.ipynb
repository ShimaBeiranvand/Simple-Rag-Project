```python
from transformers import pipeline

context = " ".join(documents)

def initializing_close_book_model(model="deepset/roberta-base-squad2"):
    qa_pipeline = pipeline("question-answering", model=model)
    return qa_pipeline

def closed_book_retriever(model , question , context):
    result = model(question=question, context=context)
    return result

```

```python
# open book method
```

```python
from datasets import load_dataset
import faiss
import numpy as np
from sentence_transformers import SentenceTransformer
from transformers import pipeline, AutoTokenizer, AutoModelForSeq2SeqLM


def initializing_open_book_model(embedder='all-MiniLM-L6-v2' , tokenizer="google/flan-t5-large" , model="google/flan-t5-large" ):

    # Embed documents
    embedder = SentenceTransformer(embedder)

    # Load generator model
    tokenizer = AutoTokenizer.from_pretrained(tokenizer)
    model = AutoModelForSeq2SeqLM.from_pretrained(model)

    return embedder , tokenizer , model


def chunk_embedder(chunk , embedder):
    doc_embeddings = embedder.encode(chunk, convert_to_tensor=False, show_progress_bar=False)
    doc_embeddings = np.array(doc_embeddings)
    return doc_embeddings

def open_book_retriever(embedded_array, documents,question, embedder, tokenizer, model, k=2, max_length=30):

    # 1. create FAISS index
    dimension = embedded_array.shape[1]
    index = faiss.IndexFlatL2(dimension)
    index.add(embedded_array)

    # 2. Embed query and retrieve top-k documents
    query_embedding = embedder.encode(question, convert_to_tensor=False)
    query_embedding = np.array(query_embedding)
    query_embedding = query_embedding.reshape(1, -1)
    distances, indices = index.search(query_embedding, k)
    retrieved_docs = [documents[i] for i in indices[0]]


    # 3. Generate answer using retrieved context
    prompt = f"question: {question} context: {retrieved_docs} answer:"
    inputs = tokenizer(prompt, return_tensors="pt", truncation=True, max_length=512)
    output = model.generate(**inputs, max_length=max_length)
    answer = tokenizer.decode(output[0], skip_special_tokens=True)

    return {'answer' : answer , 'score' : 1 / (1 + distances[0][0])}


```

```python
embedder , tokenizer , model = initializing_open_book_model()
embedded_list = []
for chunk in documents:
    embedded_list.append(chunk_embedder(chunk , embedder))
embedded_array = np.array(embedded_list)

result = open_book_retriever(embedded_array,documents , question ,embedder , tokenizer , model)
print(f"Q: {question}\nA: {result['answer']}")
```

```python
# chunking the dataset

def chunk_text(text, chunk_size=400, overlap=50):
    """
    Splits `text` into overlapping chunks of up to `chunk_size` tokens,
    with `overlap` tokens shared between consecutive chunks.
    """
    token_ids = tokenizer.encode(text, add_special_tokens=False)
    chunks = []
    start = 0
    while start < len(token_ids):
        end = start + chunk_size
        chunk_ids = token_ids[start:end]
        chunks.append(tokenizer.decode(chunk_ids, skip_special_tokens=True))
        if end >= len(token_ids):
            break
        start += chunk_size - overlap
    return chunks
```

```python
# function for giving question to every chunk and then finding the answer

def chunk_retriever(chunked_text, question , retriever, embedder=None, tokenizer=None, model= None , method=None):   # k=3

    embedded_chunks=[]
    candidates = []

    if method == 'open-book':
        for chunk in chunked_text:
            embedded_chunks.append(chunk_embedder(chunk , embedder))

        embedded_chunks = np.array(embedded_chunks)
        out = retriever(embedded_chunks,chunked_text, question, embedder, tokenizer, model)
        candidates.append({
             "answer": out["answer"],
             "score": out["score"],
             #"chunk_excerpt": chunk[:200] + "…"
        })

        # Sort candidates by score and return top_k
        candidates = sorted(candidates, key=lambda x: x["score"], reverse=True)

        return candidates

    else:
        for chunk in chunked_text:
            out = retriever(question=question, context=chunk)
            candidates.append({
               "answer": out["answer"],
               "score": out["score"],
               #"chunk_excerpt": chunk[:200] + "…"
            })

        # Sort candidates by score and return top_k
        candidates = sorted(candidates, key=lambda x: x["score"], reverse=True)

        return candidates    # return candidates[:top_k]
```

```python
import os

def create_file_content_dict(directory_path):
  """
  Creates a dictionary where keys are file names and values are file contents.

  Args:
    directory_path: The path to the directory containing the files.

  Returns:
    A dictionary where keys are file names and values are file contents.
  """

  file_content_dict = {}
  for filename in os.listdir(directory_path):
    if filename.endswith(".txt"):  # Consider only text files
      filepath = os.path.join(directory_path, filename)
      with open(filepath, 'r') as file:
        file_content_dict[filename[:-4]] = file.read().lower()
  return file_content_dict



directory_path = '/content/drive/MyDrive/astronomy dataset'
file_content_dict = create_file_content_dict(directory_path)

```

```python
real_answers = [
    'in about 23 hours and 56 minutes',
    'Jupiter is the fifth planet from the Sun',
    'Mars was formed approximately 4.5 billion years ago',
    'The surface of Mercury is heavily cratered',
    'neptune',
    'Saturns interior is thought to be composed of a rocky core, surrounded by a deep layer of metallic hydrogen',
    '82.23°',
    'because they are orbital neighbors and Venus has the most similar mass and size to Earth among the planets of the Solar System.',
    'this question shouldnt be answered',
]
```

```python
# initializing open-book and close-book methods' models

total_text = " ".join(file_content_dict.values())
chunked_text = chunk_text(total_text, chunk_size=400, overlap=50)
# context_for_open = [text for text in file_content_dict.values()]

close_model = initializing_close_book_model()
open_embedder , open_tokenizer , open_model = initializing_open_book_model()
```

```python
# creating a system of question answering without RAG

from transformers import pipeline

without_rag_pipeline = pipeline("text2text-generation", model="google/flan-t5-large")
```

```python
# checking if the answers are correct or not

def correction_checker():
    correction_check = input('Correct or not? [y/n]: ')
    if correction_check == 'y':
        return 1
    else:
        return 0

# testing the function
print(correction_checker())
```

```python
# from transformers import GPT2LMHeadModel, GPT2Tokenizer

without_rag_rating  = 0
closed_rating = 0
open_rating = 0

for question in questions:
    print('---------------------------------------------------------')

    # real answer
    print("Q: " , question)
    print('Real Answer:')
    print(real_answers[questions.index(question)])
    print('---------------------------------------------------------')

    # without rag pipeline
    input_text = f"Answer the following question factually: {question}"
    answer = without_rag_pipeline(input_text, max_length=50, do_sample=False)
    print('Without RAG:')
    print("Q: ",question)
    print("A: ",answer[0]['generated_text'])
    without_rag_rating += correction_checker()
    print('---------------------------------------------------------')

    # closed book method
    candidates = chunk_retriever(chunked_text , question , close_model, method='closed-book')
    print('Closed-Book Method')
    print("Q: " , question)
    print("A: " ,candidates[0]['answer'])
    print("Confidence Score: ", candidates[0]['score'])
    closed_rating  += correction_checker()
    print('---------------------------------------------------------')

    # open book method
    print('Open-Book Method')
    candidates = chunk_retriever(chunked_text, question , open_book_retriever, open_embedder , open_tokenizer, model =open_model , method = 'open-book' )
    print("Q: ",question)
    print("A: ",candidates[0]['answer'])
    print("Confidence Score: ", candidates[0]['score'])
    open_rating  += correction_checker()
    print('---------------------------------------------------------')
    print('---------------------------------------------------------')


```

```python
# printing the accuracy of models

print('Accuracy of the model without RAG : ')
print(without_rag_rating / len(questions) * 100)

print('\nAccuracy of the closed book method RAG : ')
print(closed_rating / len(questions) * 100)

print('\nAccuracy of the open book method RAG : ')
print(open_rating / len(questions) * 100)
```

```python
import numpy as np

def cosine_similarity(vector1, vector2):
    dot_product = np.dot(vector1, vector2)
    magnitude1 = np.linalg.norm(vector1)
    magnitude2 = np.linalg.norm(vector2)
    return dot_product / (magnitude1 * magnitude2)
```

```python
# testing related sentences

print('sentences : ')
print(sentences[4])
print(sentences[5])

similarity = cosine_similarity(embeddings_minilm[4], embeddings_minilm[5])
print("Cosine similarity:", similarity)
```

```python
# testing unrelated sentences

print('sentences : ')
print(sentences[6])
print(sentences[7])

similarity = cosine_similarity(embeddings_minilm[6], embeddings_minilm[7])
print("Cosine similarity:", similarity)
```
